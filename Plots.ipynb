```python
### PRETRAINED VS FROM SCRATCH ###
### Epoch=20 , Batch_size=16 , imgsz=640 , optimizer=Adam , lr=0.01 ###
values = {"New":74.6 , "Pretrained":76.3 }



import seaborn as sns
import matplotlib.pyplot as plt

# Data
model = list(values.keys())
mAP_values = list(values.values())

# Create bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=model, y=mAP_values, palette='muted')
plt.xlabel('Optimizers')
plt.ylabel('mAP Values')
plt.title('mAP Values for Different Models')


for i in range(len(model)):
    plt.text(i, mAP_values[i], f'{mAP_values[i]}', ha='center', va='bottom')

    
plt.ylim(0, 100)  # Optional: set y-axis limits
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: add grid lines
plt.show()
```

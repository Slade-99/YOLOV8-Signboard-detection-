```python
import os
import shutil
import pandas as pd

# Define paths
source_images_folder = "./global_images"
source_annotations_folder = "./global_annotations"
destination_data_folder = "./Dataset"
destination_images_folder = os.path.join(destination_data_folder, "Val", "Images")
destination_annotations_folder = os.path.join(destination_data_folder, "Val", "Labels")
csv_file_path = "./section1-group9/val.csv"

# Create destination folders if they don't exist
#os.makedirs(destination_data_folder, exist_ok=True)
#os.makedirs(destination_images_folder, exist_ok=True)
#os.makedirs(destination_annotations_folder, exist_ok=True)

# Read CSV file
data = pd.read_csv(csv_file_path)

# Loop through each row in CSV
for index, row in data.iterrows():
    image_filename = row['Val Images']
    annotation_filename = row['Val Annotations']

    # Construct source paths
    source_image_path = os.path.join(source_images_folder, image_filename)
    source_annotation_path = os.path.join(source_annotations_folder, annotation_filename)

    # Construct destination paths
    destination_image_path = os.path.join(destination_images_folder, image_filename)
    destination_annotation_path = os.path.join(destination_annotations_folder, annotation_filename)

    # Copy files
    shutil.copy(source_image_path, destination_image_path)
    shutil.copy(source_annotation_path, destination_annotation_path)

    print(f"Copied {image_filename} and {annotation_filename} to Data folder.")

print("Copy operation completed successfully.")
```
